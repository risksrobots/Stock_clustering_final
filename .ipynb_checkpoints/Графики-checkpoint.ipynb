```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import dates as mdates
import numpy as np
from metrics.drawdown import Drawdown_plot
```

```python
ports_real = pd.read_csv(f'df_real.csv', index_col=0)
ports_pct = pd.read_csv(f'df.csv', index_col=0)
ports_real.index = pd.to_datetime(ports_real.index)
ports_pct.index = pd.to_datetime(ports_pct.index)
```

```python
ret = pd.read_csv(f'metrics/ret_ports.csv', index_col=0)
sharp = pd.read_csv(f'metrics/sharp_ports.csv', index_col=0)
risk = pd.read_csv(f'metrics/risk_ports.csv', index_col=0)
beta = pd.read_csv(f'metrics/beta_ports.csv', index_col=0)
var = pd.read_csv(f'metrics/var.csv', index_col=0)

var.index = pd.to_datetime(var.index)
ret.index = pd.to_datetime(ret.index)
risk.index = pd.to_datetime(risk.index)
sharp.index = pd.to_datetime(sharp.index)
beta.index = pd.to_datetime(beta.index)
```

```python
ax = ports_real.plot(figsize=(15, 5))
xfmt = mdates.DateFormatter('%m,%Y')
ax.xaxis.set_major_formatter(xfmt)
plt.xlabel('Дата')# добавляем название оси Ох
plt.ylabel('Доходность')# добавляем название оси Оу
plt.title('Доходность портфелей за каждый день')# добавляем название графику
plt.legend(loc='upper left')

plt.show()
```

```python
ax = var.plot(figsize=(15, 5))

plt.xlabel('Дата')# добавляем название оси Ох
plt.ylabel('Var')# добавляем название оси Оу
plt.title('Коэффициент Var портфелей')# добавляем название графику

plt.show()
```

```python
ax = ret.plot(figsize=(15, 5))

plt.xlabel('Дата')# добавляем название оси Ох
plt.ylabel('Return')# добавляем название оси Оу
plt.title('Месячная доходность портфелей')# добавляем название графику

plt.show()
```

```python
ax = risk.plot(figsize=(15, 5))

plt.xlabel('Дата')# добавляем название оси Ох
plt.ylabel('Риск')# добавляем название оси Оу
plt.title('Коэффициент риска портфелей')# добавляем название графику

plt.show()
```
