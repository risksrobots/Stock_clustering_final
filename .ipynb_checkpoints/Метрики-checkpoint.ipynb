## Безрисковый актив

```python
df_b = pd.read_csv(r'bonds.csv', decimal=",")
df_bond = pd.DataFrame()

# Найти среднюю цену за месяц и поделить её на 100
df_bond['date'] = df_b['Дата']
df_bond['цена'] = df_b['Цена']/100
df_bond.date = pd.to_datetime(df_bond.date)
df_bond.set_index('date', inplace = True)
```

```python
df_b
```

```python
df_bond = df_bond.reset_index().groupby(pd.Grouper(key="date", freq="M")).mean()
```

```python
# =(x+1)^(1/12)-1
df_bond['цена'] = df_bond['цена'].apply(lambda x: ((x+1)**(1/12))-1)
```

## Beta портфеля

```python
var_ports = df_all_ports_pct.reset_index().groupby(pd.Grouper(key="index", freq="M")).var()
cov_ports = df_all_ports_pct.reset_index().groupby(pd.Grouper(key="index", freq="M")).cov()
```

```python
# Beta = cov(ret_port;ret_IMOEX)/var(IMOEX), beta считается, только для портфелей, не для индекса IMOEX
beta = cov_ports['sp500']/var_ports['sp500']
beta = beta.to_frame()
beta = beta.reset_index(drop=True)
beta_values = beta.sp500.values
```

## Коэффициент VaR

```python
#средний pct за месяц, стандартное отклонение за месяц
```

```python
#percentile = df_all_ports_pct.reset_index().groupby(pd.Grouper(key="date", freq="M")).quantile([0.25])
mean_for_var = df_all_ports_pct.reset_index().groupby(pd.Grouper(key="date", freq="M")).mean()
std_for_var = df_all_ports_pct.reset_index().groupby(pd.Grouper(key="date", freq="M")).std()
```

```python
l_pr_values = df_all_ports_pct.iloc[-1, :].values
```

```python
var=[]
for port in range (len(std_for_var.columns)):
    var.append(l_pr_values[port]*(1+norm.ppf(0.001, mean_for_var.iloc[:, port], std_for_var.iloc[:, port]))/l_pr_values[port]-1)
```

```python
var
```

```python
var = pd.DataFrame({'imoex':var[0], 'sharp':var[1], 'markov':var[2], 'tobin': var[3], 'traynor':var[4], 'qsharp':var[5]})
```

```python
var.index = beta_ports.index
```

```python
var
```
